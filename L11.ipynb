### 3D Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(4)
b = np.arange(4,7)
```

```python
b
```

- repeat row by elements of B
- repeat column by elements of A 

```python
a ,b = np.meshgrid(a, b)
```

```python
a
```

```python
b
```

```python
a+b
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(a, b, np.ones_like(a))
plt.show()
```

```python
a = np.arange(-1,1,0.005)
b = a
a,b = np.meshgrid(a, b)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(a, b, a**2 + b**2)
```
