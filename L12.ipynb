### visualizing iris dataset

```python
import pandas as pd
```

```python
iris = pd.read_csv("C:/Users/Kuvaladitya/Downloads/iris.csv")
```

```python
iris.head()
```

```python
iris.shape
```

```python
petal_length = iris['petal_length']
petal_width = iris['petal_width']
```

```python
plt.scatter(x=petal_length, y=petal_width)
```

```python
setosa = iris[iris['species'] == 'setosa']
verginica = iris[iris['species'] == 'virginica']
versicolor = iris[iris['species'] == 'versicolor']
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'], color='red', label='Setosa')
plt.scatter(verginica['petal_length'], verginica['petal_width'], color='blue', label='Virginica')
plt.scatter(versicolor['petal_length'], versicolor['petal_width'], color='green', label='Versicolor')	
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
```

```python
plt.hist(iris['petal_length'], bins=20)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')	
plt.show()
```

```python
iris["species"].value_counts()
```

```python
plt.pie(iris["species"].value_counts(), labels=iris["species"].unique(), autopct='%1.1f%%')
```

### Movie dataset visualization

```python
movie = pd.read_csv("C:/Users/Kuvaladitya/Downloads/movie_meta.csv")
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie['movie_title']
```

```python
def preprocess(title):
    return len(title.strip())
```

```python
movie['movie_title_len'] = movie['movie_title'].apply(preprocess)
```

```python
movie.columns
```

```python
plt.hist(movie['movie_title_len'], bins = 50)
```

### scatter way


```python
freq_title = movie['movie_title'].value_counts()
freq_title
```

```python
freq_title = {}

for l in movie['movie_title_len']:
	if l in freq_title:
		freq_title[l] += 1
	else:
		freq_title[l] = 1
```

```python
freq_title
```

```python
x = list(freq_title.keys())
y = list(freq_title.values())
plt.scatter(x, y)
plt.xlabel('Movie Title Length')
plt.ylabel('Frequency(number of movies)')
plt.show()
```
