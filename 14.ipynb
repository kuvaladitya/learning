### Face Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('C:/Users/Kuvaladitya/Downloads/group-photo.jpg')
```

```python
detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
all_faces = detector.detectMultiScale(img, 1.2, 5)
```

```python
all_faces = detector.detectMultiScale(img)
```

```python
all_faces.shape
```

```python
all_faces[0]
```

```python
x, y, w, h = all_faces[0]
```

```python
img_copy = img.copy()
x, y, w, h = all_faces[0]
cv2.rectangle(img_copy, (x, y), (x + w, y + h), (0, 0, 0), 2)
```

```python
plt.imshow(img_copy)
plt.axis('off')
plt.show()
```

```python
for face in all_faces:
    x,y,w,h = face
    img = cv2.rectangle(img_copy, (x, y), (x + w, y + h), (255, 0, 0), 2)
    
plt.imshow(img)
```
