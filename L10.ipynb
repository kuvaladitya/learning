### Bar graph
- plt.bar
- plot a bar graph between high and labels you provide

```python
x = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
plt.bar(x, car1, width=0.3, tick_label=['2017', '2018', '2019', '2020'], label='car1')
plt.bar(x + 0.3, car2, width=0.3, label='car2')
plt.legend()
plt.xlabel("year")
plt.ylabel("sales in thousands")
plt.title("Car Sales")
plt.xlim(-0.5, 4)
```

### Pie Chart
- plt.pie
- plot the pie chart with the provided labels and  values

```python
marks = np.array([78, 45, 90, 65])
plt.pie(marks, labels=['physics', 'chemistry', 'maths', 'english'], explode=[0.1, 0, 0, 0], autopct='%0.2F')
```

### Histograms
- plt.hist()


```python
x_snd = np.random.randn(500)
```

```python
np.round(x_snd*8 + 70)
```

```python
maths_mu = 70
maths_sigma = 8
maths_marks =x_snd*maths_sigma + maths_mu
np.round(maths_marks)

phy_mu = 50
phy_sigma = 10
phy_marks = x_snd*phy_sigma + phy_mu
np.round(phy_marks)
```

```python
plt.hist(maths_marks, bins=10)
plt.hist(phy_marks, bins=10, alpha=0.5)
plt.hist
```

### Subplots
- create multiple plots in just a single figure

```python
x = np.arange(0, 10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, y1)

plt.subplot(2, 3, 2)
plt.plot(x, y2)

plt.subplot(2, 3, 3)
plt.plot(x, y3)

plt.subplot(2, 3, 4)
plt.plot(x, y4)

plt.subplot(2, 3, 5)
plt.plot(x, y3)

plt.subplot(2, 3, 6)
plt.plot(x, y4)

plt.show()
```
