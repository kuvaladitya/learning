```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:/Users/Kuvaladitya/Downloads/dog.jpg")
```

```python
plt.imshow(img)
```

### Changing ColorSpace

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_hsv)
```

### Image Resizing

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
img_new = cv2.resize(img_rgb,(500,500))
```

```python
plt.imshow(img_new)
```

### blur

```python
img_blur = cv2.blur(img_new , (10,10))
plt.imshow(img_blur)
```

### Image Rotation

```python
img_rgb.shape
```

```python
rows,cols = img_rgb.shape[:-1]
```

```python
M = cv2.getRotationMatrix2D((rows/2,cols/2),-90,1)
```

```python
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
```

### Image Translation

```python
M = np.float32(([1,0.1,50],[0,1,-200]))
```

```python
M
```

```python
i = cv2.warpAffine(img_rgb,M,(cols,rows))
```

```python
plt.imshow(i)
```

### edge detection
- canny edge detection

```python
i = cv2.Canny(img_rgb, 50, 50)
plt.imshow(i)
plt.axis('off')
```
